```python
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
geolocator = Nominatim(user_agent="walter.tyrna@gmail.com")
import pandas as pd
```

```python
data = pd.read_csv('clean_data.csv')
```

```python
locs = []
lat = []
lon = []
for i in data.coordinates:
    i = str(i)
    i = i.replace('(','')
    i = i.replace(')','')
    i = i.split(',')
    lats = i[0]
    lons = i[1].replace(' ','')
    locs.append(i)
    lat.append(lats)
    lon.append(lons)
```

```python
data['lat'] = lat
data['lon'] = lon
```

```python
map = folium.Map(location=[38.901918699999996, -77.03895284983707], zoom_start=14, control_scale=True)
```

```python
map
```

```python
for index, location_info in data.iterrows():
    folium.CircleMarker([location_info["lat"], location_info["lon"]], radius=2, weight=1, color='red', fill_color='red', fill_opacity=.5 ).add_to(map)
```
